## Automated Preprocessing with Sklearn Pipeline

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

## Step 1 : Read the datasets

```python
import pandas as pd
df = pd.read_csv('Cars93.csv', na_values=['', 'NA'], keep_default_na=False)
df.head()
```

## Step 2  : Perform basic data quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m
```

```python
m[m>0]
```

```python
df.duplicated().sum()
```

## Drop the duplicates

```python
df = df.drop_duplicates(keep='first')
```

## Step 3 : Seperate X and Y (Weight)

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

## Step 4: Automated preprocessing

```python

# Seperate cat and con features for X 
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline # Automating Preprocessing
from sklearn.impute import SimpleImputer # Replacing missing values
from sklearn.preprocessing import StandardScaler, OneHotEncoder # Preprocessing 
from sklearn.compose import ColumnTransformer # To combine numerical and categorical columns
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
pre
```

## Apply pre.fit_transform on X

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
X_pre.to_csv('preprocessed.csv', index=False)
```

## Step 5 : Perform train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2,
                                                random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

## step 6 : build the model


```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

## Evaluate r2 scores in train and test


```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

## Get MSE, RMSE, MAE, R2

```python
def evaluate_model(model, x, y):
    # Predict the results for x
    ypred = model.predict(x)
    # Import packages
    from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print mse, rmse, mae, r2
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(model, xtrain, ytrain)
```

```python
evaluate_model(model, xtest, ytest)
```
